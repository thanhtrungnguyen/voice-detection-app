```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
audio_file = r'C:\Users\thanh\OneDrive\Documents\Sound Recordings\Recording (2).wav'  # Change this to your audio file path
data, sample_rate = sf.read(audio_file)

# Parameters for voice activity detection
frame_length = 2048
hop_length = 512
threshold = 0.02  # Adjust this threshold based on your needs

# Calculate the short-time energy of the audio signal
energy = np.array([
    np.sum(np.abs(data[i:i + frame_length] ** 2))
    for i in range(0, len(data), hop_length)
])

# Detect voice activity based on energy threshold
voice_activity = energy > threshold

# Generate time axis for plotting
time = np.linspace(0, len(data) / sample_rate, len(voice_activity))

# Plotting the results
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(np.arange(len(data)) / sample_rate, data, color='b')
plt.title('Audio Signal Waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(time, voice_activity.astype(int), color='r')  # Cast boolean to int for plotting
plt.fill_between(time, voice_activity.astype(int), color='lightcoral', alpha=0.5)
plt.title('Voice Activity Detection')
plt.xlabel('Time [s]')
plt.ylabel('Voice Activity (1=Voice, 0=Silence)')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
audio_file = r'C:\Users\thanh\OneDrive\Documents\Sound Recordings\Recording (2).wav'  # Change this to your audio file path
data, sample_rate = sf.read(audio_file)

# Parameters for voice activity detection
frame_length = 2048
hop_length = 512
threshold = 0.02  # Adjust this threshold based on your needs

# Calculate the short-time energy of the audio signal
energy = np.array([
    np.sum(np.abs(data[i:i + frame_length] ** 2))
    for i in range(0, len(data), hop_length)
])

# Detect voice activity based on energy threshold
voice_activity = energy > threshold

# Generate time axis for plotting
time = np.linspace(0, len(data) / sample_rate, len(voice_activity))

# Create a combined plot
plt.figure(figsize=(12, 6))

# Plot the audio signal waveform
plt.plot(np.arange(len(data)) / sample_rate, data, color='b', label='Audio Signal')
plt.title('Audio Signal Waveform with Voice Activity Detection')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()

# Overlay voice activity detection
plt.fill_between(time, np.min(data), np.max(data), where=voice_activity, 
                 color='lightcoral', alpha=0.5, label='Voice Activity')

plt.legend()
plt.tight_layout()
plt.show()
```
